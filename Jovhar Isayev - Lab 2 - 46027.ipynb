```python
import requests as req;
```

```python
from bs4 import BeautifulSoup as bsoup;
```

```python
page = req.get('https://www.imdb.com/chart/boxoffice');
```

```python
print(page.status_code);
```

```python
soup = bsoup(page.content,'html.parser');
```

```python
for movie in soup.find_all('tr'):
    name = movie.getText();
    print(name);
```
